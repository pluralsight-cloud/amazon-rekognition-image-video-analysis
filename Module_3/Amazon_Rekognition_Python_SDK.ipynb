## Image Analysis Tasks Using Amazon Rekognition APIs

### Request Format
To know more about the request format for the detect_labels API, visit this link: https://docs.aws.amazon.com/rekognition/latest/APIReference/API_DetectLabels.html

```python
{
   "Features": [ "string" ], # Valid values: GENERAL_LABELS, IMAGE_PROPERTIES
   "Image": { 
      "Bytes": blob,
      "S3Object": { 
         "Bucket": "string",
         "Name": "string",
         "Version": "string"
      }
   },
   "MaxLabels": number,
   "MinConfidence": number,
   "Settings": { 
      "GeneralLabels": { 
         "LabelCategoryExclusionFilters": [ "string" ],
         "LabelCategoryInclusionFilters": [ "string" ],
         "LabelExclusionFilters": [ "string" ],
         "LabelInclusionFilters": [ "string" ]
      },
      "ImageProperties": { 
         "MaxDominantColors": number
      }
   }
}
```

### Input Image
![image](M3_C1_Image_Object_Scenes.jpg)

```python
import boto3
rekognition_client = boto3.client('rekognition')
```

```python
with open('M3_C1_Image_Object_Scenes.jpg', 'rb') as image_file:
    image_bytes = image_file.read()
```

```python
detect_label_response = rekognition_client.detect_labels(
    Image={
        'Bytes': image_bytes
    },
    MaxLabels=10,
    MinConfidence=70
)
```

```python
detect_label_response
```

```python
with open('M3_C1_Image_Content_Moderation.jpg', 'rb') as image_file:
    image_bytes = image_file.read()
```

```python
detect_moderation_label_response = rekognition_client.detect_moderation_labels(
    Image={
        'Bytes': image_bytes
    },
    MinConfidence=70
)
```

```python
detect_moderation_label_response
```

```python
with open('M3_C1_Image_Text_Detection.jpg', 'rb') as image_file:
    image_bytes = image_file.read()
```

```python
detect_text_response = rekognition_client.detect_text(
    Image={
        'Bytes': image_bytes
    },
)
```

## API Call Type

**Synchronous**: When a synchronous API call is made, the calling program waits for the API to return a response before proceeding to the next line of code.

**Asynchronous**: When an asynchronous API call is made, the calling program does not wait for the API to return a response. It submits the job and continues executing other tasks while the API call is processed in the background. Typically, there exists another API call that can be used to check the status of the job and get the response once the job is finished.

**StartLabelDetection** API: https://docs.aws.amazon.com/rekognition/latest/APIReference/API_StartLabelDetection.html

**GetLabelDetection** API: https://docs.aws.amazon.com/rekognition/latest/APIReference/API_GetLabelDetection.html

## Video Label Detection

```python
start_label_detection_response = rekognition_client.start_label_detection(
    Video={
        'S3Object': {
            'Bucket': 'rekognition-bucket-1000',
            'Name': 'input-output/M3_C3_Video_Analysis.mp4'
        }
    },
    MinConfidence = 90,
    # NotificationChannel={
    #     'SNSTopicArn': 'string',
    #     'RoleArn': 'string'
    # }
)
```

```python
start_label_detection_response
```

```python
jobid = start_label_detection_response['JobId']
jobid
```

**GetLabelDetection** parameters: https://docs.aws.amazon.com/rekognition/latest/APIReference/API_GetLabelDetection.html

```python
get_label_detection_response = rekognition_client.get_label_detection(
    JobId=jobid,
    # MaxResults=123,
    # NextToken='string',
    # SortBy='NAME'|'TIMESTAMP',
    # AggregateBy='TIMESTAMPS'|'SEGMENTS'
)
```
